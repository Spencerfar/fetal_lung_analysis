```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df15 = pd.read_csv('TP_radius_scores_TP_GW15_smaller.csv', index_col=0).T
df18 = pd.read_csv('TP_radius_scores_TP_GW18_smaller.csv', index_col=0).T
df18.rename(index={'Early AT2-like cells':'Tip cells'},inplace=True)
```

```python
selected_celltypes = ['Mature ciliated cells',
'Basal cells',
'Airway SMC',
'Airway fibroblast progenitors',
'Cycling fibroblasts',
'SMG secretory cells',
'Early fibroblasts',
'Vascular SMC',
'Lipofibroblast precursors',
'SMG basal cells',
'Cycling early capillaries',
'gCap',
'Tip cells',
'Intermediate lymphatic EC',
'NKX2-1+SOX9+CFTR+ cells',
'KIT+CA4+ endothelial cells',
'PTEN+STAT3+ cells',
'SOX2highCFTR+ cells',
'Lymphatic EC',
'Arterial EC',
'Venous EC',
#'SCGB3A2+SFTPB+CFTR+ cells',
'PNEC',
'SCGB3A2+FOXJ1+ cells',
'Club cells',
'NRGN+ cells',
'Erythrocyte-like EC-1',
'Chondrocyte',
'SOX2lowCFTR+ cells',
'Proliferating progenitors',
'Ciliated precursor cells',
'Aerocyte',
'Lipofibroblasts',
'Budtip progenitors',
'Uncommitted cells',
'Unknown stromal cells',
'Erythrocyte-like EC-2']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(4,5))
g = sns.heatmap(df15.loc[selected_celltypes],
           yticklabels=True,ax=ax, cmap='Spectral_r',#vmax=30,
                cbar_kws={'label': 'Co-occurrence proportion'},linewidths=0.5, linecolor='#D3D3D3',
               vmax=0.3, vmin=0)
g.figure.axes[-1].yaxis.label.set_size(9)
g.figure.axes[-1].set_yticklabels(g.figure.axes[-1].get_ymajorticklabels(), fontsize = 8)
ax.set_title('GW 15', size=10)
ax.set_xlabel('Radius from SCGB3A2+SFTPB+CFTR+ cell', size=9)
ax.set_ylabel('Deconvolved celltype', size=9)
ax.set_xticklabels(df15.columns.values, rotation=0, size=8)
#ax.tick_params(axis='y',size=8)
ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 8)
#g.set_ylabel('SCGB3A2+SFTPB+CFTR+ co-occurrence ratio')
plt.tight_layout()
#plt.savefig('figures/TP_proximity_plot_GW15.pdf', bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(4,5))
g = sns.heatmap(df18,
           yticklabels=True,ax=ax, vmin=0, vmax=0.3,cmap='Spectral_r',#vmax=30,
                cbar_kws={'label': 'Co-occurrence proportion'},linewidths=0.5, linecolor='#D3D3D3')
g.figure.axes[-1].yaxis.label.set_size(9)
g.figure.axes[-1].set_yticklabels(g.figure.axes[-1].get_ymajorticklabels(), fontsize = 8)
ax.set_title('GW 18', size=10)
ax.set_xlabel('Radius from SCGB3A2+SFTPB+CFTR+ cell', size=9)
ax.set_ylabel('Deconvolved celltype', size=9)
ax.set_xticklabels(df18.columns.values, rotation=0, size=8)
#ax.tick_params(axis='y',size=8)
ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 8)
#g.set_ylabel('SCGB3A2+SFTPB+CFTR+ co-occurrence ratio')
plt.tight_layout()
#plt.savefig('figures/TP_proximity_plot_GW18.pdf', bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,2, figsize=(5,5))
ax=ax.flatten()

g = sns.heatmap(df15.loc[selected_celltypes][[25,50,75,100]],
           yticklabels=True,ax=ax[0], vmin=0, vmax=0.3, cmap='Spectral_r',#vmax=30,
                cbar_kws={'label': 'Co-occurrence proportion'},linewidths=0.5, linecolor='#D3D3D3')

g.figure.axes[-1].set_visible(False)
ax[0].set_title('GW 15', size=11)
ax[0].set_yticklabels(ax[0].get_ymajorticklabels(), fontsize = 8)

g = sns.heatmap(df18.loc[selected_celltypes][[25,50,75,100]],
           yticklabels=False,ax=ax[1], vmin=0, vmax=0.3, cmap='Spectral_r',#vmax=30,
                cbar_kws={'label': 'Co-occurrence proportion'},linewidths=0.5, linecolor='#D3D3D3')
g.figure.axes[-1].yaxis.label.set_size(9)
g.figure.axes[-1].set_yticklabels(g.figure.axes[-1].get_ymajorticklabels(), fontsize = 8)
ax[1].set_title('GW 18', size=11)
ax[0].set_xlabel('                              Radius from SCGB3A2+SFTPB+CFTR+ cell', size=9)
ax[0].set_ylabel('Deconvolved celltype', size=9)
ax[0].set_xticklabels(df18[[25,50,75,100]].columns.values, rotation=0, size=8)
ax[1].set_xticklabels(df18[[25,50,75,100]].columns.values, rotation=0, size=8)
ax[0].tick_params(width=0.5)
ax[1].tick_params(width=0.5)
#ax.tick_params(axis='y',size=8)
#ax[1].set_yticklabels([''], fontsize = 8)

plt.tight_layout()
plt.savefig('figures/TP_proximity_plot_combined.pdf', bbox_inches='tight')
plt.show()
```
